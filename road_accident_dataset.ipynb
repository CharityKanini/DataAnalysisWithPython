```python
import pandas as pd

# Load the dataset
data = pd.read_csv('road_accident_dataset.csv')

# Copy the data to a new DataFrame
df = data.copy()

# Display the first few rows of the DataFrame
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 5))
sns.countplot(x=df['Year'], color='skyblue')
plt.title("Accidents by Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
colors = ['blue', 'red', 'green', 'orange' , 'purple']
df['Weather Conditions'].value_counts().plot.pie(autopct="%1.1f%%", colors=colors,startangle=90, wedgeprops={'edgecolor': 'black'})
plt.title("Accidents by Weather Conditions", fontsize=14, fontweight='bold')
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x=df['Day of Week'], order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], palette="Set2")
plt.title("Accidents by Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
colors = ['blue', 'green']
sns.countplot(x=df["Urban/Rural"], palette=colors)
plt.title("Accidents in Urban vs Rural Areas", fontsize=14, fontweight='bold')
plt.xlabel("Location", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['Number of Vehicles Involved'], bins=10, kde=True, color="crimson")
plt.title("Number of Vehicles Involved in Accidents", fontsize=14, fontweight='bold')
plt.xlabel("Number of Vehicles", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))


region_colors = ['blue', 'green', 'red', 'purple', 'orange']

sns.countplot(y=df['Region'], order=df['Region'].value_counts().index, palette=region_colors)
plt.title("Accidents by Region")
plt.xlabel("Count")
plt.ylabel("Region")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x=df["Visibility Level"], color='skyblue')  # Custom color for the boxplot
plt.title("Visibility Level Distribution in Accidents", fontsize=14, fontweight='bold')
plt.xlabel("Visibility Level", fontsize=12)
plt.show()
```

```python

plt.figure(figsize=(10, 5))
sns.scatterplot(x=df["Traffic Volume"], y=df["Number of Vehicles Involved"], alpha=0.7, color="purple")
plt.title("Traffic Volume vs Number of Vehicles Involved", fontsize=14, fontweight='bold')
plt.xlabel("Traffic Volume", fontsize=12)
plt.ylabel("Number of Vehicles Involved", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 5))

yearly_fatalities = df.groupby('Year').size().reset_index(name='Fatalities')

# Plot the line plot
sns.lineplot(x='Year', y='Fatalities', data=yearly_fatalities, marker='o', color='blue')

plt.title("Number of Fatalities Over the Years", fontsize=14, fontweight='bold')
plt.xlabel("Year", fontsize=12)
plt.ylabel("Number of Fatalities", fontsize=12)
plt.show()
```
